```python
#importing required library

import numpy as np # linear algebra

import pandas as pd # data processing CSV file I/O

import matplotlib.pyplot as plt
```

```python
# Next, we'll load the train and test dataset,

train = pd.read_csv('train_santander.csv')

test =  pd.read_csv("test_santander.csv")

sample = pd.read_csv("sample_submission.csv")
```

```python
train.head()
```

```python
# A little less then 4% are unhappy => unbalanced dataset

df = pd.DataFrame(train.TARGET.value_counts())

df['Percentage'] = 100*df['TARGET']/train.shape[0]

df
```

```python
sample.shape
```

```python
testing_data = pd.merge(test,sample)
```

```python
testing_data.head()
```

```python
new_data = pd.concat([testing_data,train])
```

```python
new_data.head() 
```

```python
new_data.drop(new_data.columns[0],axis=1,inplace=True) # dropping 'id' coluumn from the new dataset
```

```python
new_data.shape
```

```python
#creating the labels for dataset

labels = np.array(new_data['TARGET']).reshape(new_data['TARGET'].shape[0],1) #reshape column="TARGET"
```

```python
#converting fthe dataset into array

x =np.array(new_data)

print(x)
```

```python
#Principal Component Analysis (PCA) identifies the combination of attributes (principal components, or directions in the feature space)
    
from sklearn.decomposition import PCA #absolute import ~ sklearn.decomposition
    
pca = PCA(n_components = 170) 

features = pca.fit_transform(x)
    
```

```python
## splitting the data into training and testing sets

from sklearn.model_selection import train_test_split  

x_train, x_test, y_train, y_test = train_test_split(features,labels,train_size = 0.7, test_size = 0.3, random_state = 0)
```

```python
# Fitting Naive Bayes to the Training set  

from sklearn.naive_bayes import GaussianNB 

model = GaussianNB()
```

```python
# Train the model using the training sets

model.fit(x_train,y_train)
```

```python
#accuracy_score after fitting naive bayes to the training set

from sklearn.metrics import accuracy_score 

#Predict the response for test dataset

y_pred = model.predict(y_test)
```

```python
y_pred
```

```python
# Model accuracy ~~~ y_test

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report #defining the final classification report 

print(classification_report(y_pred,y_test))
```
